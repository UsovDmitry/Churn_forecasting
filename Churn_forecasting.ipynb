"Оцифруем" распределения + выведем на гистограммы.

```python
def exited_ratio(features,title,xl,yl,kde=False):
    ft = df.groupby([features, 'Exited'])['RowNumber'].count().rename('%%')
    b = ft /  ft.groupby(level=0).sum() 
    df_ = pd.DataFrame(data=100*b)
    df_['count'] = ft
    
    fig, ax = plt.subplots(figsize=(7,6))
    sns.histplot(data=df, x=features,hue='Exited',kde=kde ) 
    plt.legend(['Ушел из Банка', 'Клиент Банка'], prop={'size':12})
    plt.title(title)
    plt.xlabel(xl)
    plt.ylabel(yl)
    
    return df_
```

```python
def tenure_category(row):
    tenure_category = row['Tenure']
    
    if ((tenure_category>=0) & (tenure_category<=3)):
        return 1
    elif ((tenure_category>=4) & (tenure_category<=7)):
        return 2
    else:
        return 3

df['tenure_category'] = df.apply(tenure_category, axis=1)


drop_columns = ['RowNumber','CustomerId', 'Surname']
df_upg = df.drop(drop_columns, axis=1)
df_upg.head()
```

```python
df_upg = pd.get_dummies(data=df_upg, columns=['Geography','Gender'], drop_first=True)
df_upg.head()
```

```python
df_train = df_upg[~df_upg['Tenure'].isna()]
df_train.head()
```

```python
df_test = df_upg[df_upg['Tenure'].isna()]
df_test['Tenure'] = df_test['Tenure'].fillna(0)
df_test.head()
```

```python
drop_columns = ['Exited','Geography_Germany','Geography_Spain','Gender_Male'] #'HasCrCard','EstimatedSalary',
df_train = df_train.drop(drop_columns, axis=1)
df_train.head()
```

```python
X_train = df_train.drop(['Tenure','tenure_category'], axis=1)
Y_train = df_train['tenure_category']

features_train, features_valid, target_train, target_valid = train_test_split(
    X_train, Y_train, test_size=0.25, random_state=12345)
```

```python
model = RandomForestClassifier(random_state=123451, n_estimators = 45)
model_fit = model.fit(features_train,target_train)
result = model_fit.score(features_train,target_train) 
result
```

```python
pred_valid=model_fit.predict(features_valid)
result = model_fit.score(features_valid,target_valid) 
print (result)
accuracy = accuracy_score(target_valid,pred_valid)
print (accuracy)
```

### Борьба с дисбаллансом, техника upsample

Один из методов борьбы с дисбалансом, это применение техники upsampling - когда объекты редкого класса становяться соизмеримы с размером объектов основного класса. В данной ситуации, необходимо увеличить кол-во объектов принадлежащих классу "1".
Напишем ниже функцию для этого.

```python
def upsample(features,target,repeat):
    
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    features_upsampled, target_upsampled = shuffle(features_upsampled, target_upsampled,random_state=12345)

    return features_upsampled, target_upsampled
```

### Борьба с дисбаллансом, техника downsample

Другой методов борьбы с дисбалансом, это применение техники downsampling- когда объекты "частого" класса становяться соизмеримы с размером объектов  не основного класса. В данной ситуации, необходимо уменьшить кол-во объектов принадлежащих классу "0".
Реализуем это ниже в виде функции.

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled

features_downsampled, target_downsampled = downsample(features_train, target_train, 0.1)
features_valid_downsampled, target_valid_downsampled = downsample(features_valid, target_valid, 0.26)
```
